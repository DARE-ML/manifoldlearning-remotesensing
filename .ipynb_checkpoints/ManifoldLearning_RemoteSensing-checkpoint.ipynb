### Loading and visualising the dataset

```python
# import the data
data_array = loadmat('./Indian_pines_corrected.mat')['indian_pines_corrected'] # hyperspectral data
gt = loadmat('./Indian_pines_gt.mat')['indian_pines_gt'] # ground truth data
```

```python
# visualisation (read and enhance)
def plot_array_stretched(array, label, n):
    fig = plt.figure(figsize=(8, 6), dpi=100)
    plt.axis('off')
    v_mean = np.array(np.nanmean(array))
    v_std = np.array(np.nanstd(array))
    cb = plt.imshow(array, cmap='Greys_r', vmin=v_mean-(n*v_std), vmax=v_mean+(n*v_std))
    plt.colorbar(cb, shrink=1, label=label)
    plt.show()

@interact(band=range(1, data_array.shape[2]+1))
def show_dist(band):
    plot_array_stretched(data_array[:, :, band-1], f'Band {band}', 2)
```

```python
# plot the ground truth data
def plot_data(data, n_clusters):
    fig = plt.figure(figsize=(8, 6), dpi=100)
    plt.axis('off')
    # cmap = cm.get_cmap('nipy_spectral', n_clusters)
    cmap = cm.get_cmap('tab20c', n_clusters)
    plt.imshow(data, cmap=cmap)
    cbar = plt.colorbar(label='Cluster Label')
    tick_locs = (np.arange(n_clusters) + 0.5)*(n_clusters-1)/n_clusters
    cbar.set_ticks(tick_locs)
    cbar.set_ticklabels(np.arange(n_clusters))
    
    plt.show()

plot_data(gt, 17)
```

```python
# reshape the input data from rcb to samples and features
data_reshaped = data_array.reshape(data_array.shape[0]*data_array.shape[1], -1)
data_reshaped = minmax_scale(data_reshaped, feature_range=(0, 1), axis=0, copy=False)
gt_reshaped = gt.flatten()
```

### Principal Component Analysis (PCA)

```python
# define the model
model = PCA(n_components=2, random_state=1)
# define the space of hyperparameters to search
# whiten: enable or disable whitening
# svd_solver: algorithm used for singular value decomposition
search_space = [Categorical([True, False], name='whiten'),
                Categorical(['auto', 'full', 'arpack', 'randomized'], name='svd_solver')]
# create an object of the classifier
clusterer = cluster.KMeans(n_clusters=17, random_state=1)

# define the function used to evaluate a given configuration
@use_named_args(search_space)
def evaluate_model(**params):
    # setting the parameters of the model
    model.set_params(**params)
    # fitting the model
    selected_components = model.fit_transform(data_reshaped)
    labels = clusterer.fit_predict(selected_components)
    result= v_measure_score(gt_reshaped, labels)
    return 1.0 - result

# perform optimisation
result = gp_minimize(evaluate_model, search_space, n_jobs=-1, n_calls=100)
# summarising finding
print(f'Best V-measure: {1.0 - result.fun}')
print(f'Best Parameters: whiten = {result.x[0]}, svd_solver = {result.x[1]}')

# Best V-measure: 0.28820604179631726
# Best Parameters: whiten = False, svd_solver = auto
```

```python
# PCA
pca = PCA(n_components=2, random_state=1)
selected_components_pca = pca.fit_transform(data_reshaped)

# k-means clustering using the selected pricipal components
clusterer = cluster.KMeans(n_clusters=17, random_state=1) # create an object of the classifier
clusterer.fit(selected_components_pca)
labels = clusterer.predict(selected_components_pca)
labels_2d = labels.reshape(data_array[:, :, 0].shape) # reshape the labels to a 3D array (single band)
plot_data(labels_2d, 17)
```

```python
fig = plt.figure(figsize=(8, 6), dpi=100)

n_clusters = 16
# colors = cm.nipy_spectral(labels / n_clusters)
colors = cm.tab20c(labels / n_clusters)

plt.scatter(
    selected_components_pca[:, 0],
    selected_components_pca[:, 1],
    marker='.', s=30, lw=0, alpha=0.7, c=colors, edgecolor='k'
)

# labelling the clusters
centers = clusterer.cluster_centers_

# draw white circles at cluster centres
plt.scatter(
    centers[:, 0],
    centers[:, 1],
    marker='o', s=200, alpha=1, c='white', edgecolor='k'
)

for i, c in enumerate(centers):
    plt.scatter(c[0], c[1], marker='$%d$' %i, s=50, alpha=1, edgecolor='k')

plt.xlabel('1st Component')
plt.ylabel('2nd Component')

plt.show()
```

```python
# colors = cm.nipy_spectral([x/n_clusters for x in list(range(n_clusters+1))])
colors = cm.tab20c([x/n_clusters for x in list(range(n_clusters+1))])

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

plot_cluster_cardinality(clusterer.labels_, ax=ax1, title='Cardinality', color=colors)
plot_cluster_magnitude(selected_components_pca, clusterer.labels_, clusterer.cluster_centers_, euclidean, ax=ax2,
                       title='Magnitude', color=colors)
plot_magnitude_vs_cardinality(selected_components_pca, clusterer.labels_, clusterer.cluster_centers_, euclidean,
                              ax=ax3, title='Magnitude vs. Cardinality', color=colors)

fig.autofmt_xdate(rotation=0)
plt.tight_layout()
plt.show()
```

```python
features = ['1st Component', '2nd Component']
selected_components_pca_df = pd.DataFrame(selected_components_pca, columns=features)
selected_components_pca_df['cluster'] = clusterer.labels_

ncols = 2
nrows = len(features) // ncols + (len(features) % ncols > 0)

fig = plt.figure(figsize=(10, 5))

for n, feature in enumerate(features):
    ax = plt.subplot(nrows, ncols, n+1)
    box = selected_components_pca_df[[feature, 'cluster']].boxplot(by='cluster', ax=ax, return_type='both', patch_artist=True)

    for row_key, (ax, row) in box.iteritems():
        ax.set_xlabel('Cluster Label')
        ax.set_title('')
        ax.set_ylabel(feature)
        ax.grid(False)
        
        for i, box_ in enumerate(row['boxes']):
            box_.set_facecolor(colors[i])

fig.suptitle('')
plt.tight_layout()
plt.show()
```

```python
# calculate different performance metrics for various numbers of clusters
range_n_clusters = list(range(15, 20))
columns = ['Number of clusters', 'Average Silhouette score', 'Calinski-Harabasz score', 'Davies-Bouldin score',
           'Homogeneity score', 'Completeness score', 'V-measure', 'Rand index', 'Adjusted Rand index',
           'Mutual Information score', 'Adjusted Mutual Information score']
scores_pca = pd.DataFrame(np.zeros((len(range_n_clusters), len(columns))), columns=columns)

# 3D plots and different scores
for count, n_clusters in enumerate(range_n_clusters):
    # initialise the clusterer with n_clusters value and a random generator
    clusterer = cluster.KMeans(n_clusters=n_clusters, random_state=1)
    labels = clusterer.fit_predict(selected_components_pca)

    # the silhouette_score gives the average value for all the samples
    # this gives a perspective into the density and separation of the formed clusters
    silhouette_avg = silhouette_score(selected_components_pca, labels)
    
    scores_temp = [n_clusters, silhouette_avg,
                   # the resulting Calinski-Harabasz score
                   calinski_harabasz_score(selected_components_pca, labels),
                   # the resulting Davies-Bouldin score
                   # the minimum score is zero, and the lower values the better clustering performance
                   davies_bouldin_score(selected_components_pca, labels),
                   # homogeneity score between 0.0 and 1.0. 1.0 stands for perfectly homogeneous labelling
                   homogeneity_score(gt_reshaped, labels),
                   # completeness score between 0.0 and 1.0. 1.0 stands for perfectly complete labelling
                   completeness_score(gt_reshaped, labels),
                   # v-measure is the harmonic mean of the first two
                   v_measure_score(gt_reshaped, labels),
                   # similarity score between 0.0 and 1.0, inclusive, 1.0 stands for perfect match
                   rand_score(gt_reshaped, labels),
                   # similarity score between -1.0 and 1.0
                   # random labellings have an ARI close to 0.0. 1.0 stands for perfect match
                   adjusted_rand_score(gt_reshaped, labels),
                   # mutual information, a non-negative value, measured in nats using the natural logarithm
                   mutual_info_score(gt_reshaped, labels),
                   # the AMI returns a value of 1 when the two partitions are identical (ie perfectly matched)
                   # random partitions (independent labellings) have an expected AMI around 0 on average hence can be negative
                   # the value is in adjusted nats (based on the natural logarithm)
                   adjusted_mutual_info_score(gt_reshaped, labels)
                   ]
    scores_pca.iloc[[count]] = scores_temp
    
from IPython.display import display, HTML
display(HTML(scores_pca.to_html(index=False)))
```

### t-Distributed Stochastic Neighbour Embedding (t-SNE)

```python
# define the model
model = TSNE(random_state=1, n_jobs=-1)
# define the space of hyperparameters to search
# perplexity: controls the balance between preserving global and local structures in the data
# learning_rate: determines the step size used in the optimisation process
search_space = [Integer(5, 50, name='perplexity'),
                Integer(10, 1000, name='learning_rate')]
# create an object of the classifier
clusterer = cluster.KMeans(n_clusters=17, random_state=1)

# define the function used to evaluate a given configuration
@use_named_args(search_space)
def evaluate_model(**params):
    # setting the parameters of the model
    model.set_params(**params)
    # fitting the model
    selected_components = model.fit_transform(data_reshaped)
    labels = clusterer.fit_predict(selected_components)
    result= v_measure_score(gt_reshaped, labels)
    return 1.0 - result

# perform optimisation
result = gp_minimize(evaluate_model, search_space, n_jobs=-1, n_calls=100)
# summarising finding
print(f'Best V-measure: {1.0 - result.fun}')
print(f'Best Parameters: perplexity = {result.x[0]}, learning_rate = {result.x[1]}')

# Best V-measure: 0.30531505207626086
# Best Parameters: perplexity = 50, learning_rate = 389
```

### Spectral Embedding (Laplacian Eigenmaps)

```python
# define the model
model = SpectralEmbedding(random_state=1, n_jobs=-1)
# define the space of hyperparameters to search
# affinity: determines the method used to construct the affinity matrix, which captures the relationship between data points
# gamma: controls the scaling factor for the affinity matrix
search_space = [Categorical(['nearest_neighbors', 'rbf'], name='affinity'),
                Real(0.1, 1, name='gamma')]
# create an object of the classifier
clusterer = cluster.KMeans(n_clusters=17, random_state=1)

# define the function used to evaluate a given configuration
@use_named_args(search_space)
def evaluate_model(**params):
    # setting the parameters of the model
    model.set_params(**params)
    # fitting the model
    selected_components = model.fit_transform(data_reshaped)
    labels = clusterer.fit_predict(selected_components)
    result= v_measure_score(gt_reshaped, labels)
    return 1.0 - result

# perform optimisation
result = gp_minimize(evaluate_model, search_space, n_jobs=-1, n_calls=100)
# summarising finding
print(f'Best V-measure: {1.0 - result.fun}')
print(f'Best Parameters: affinity = {result.x[0]}, gamma = {result.x[1]}')

# Best V-measure: 0.342319671523847
# Best Parameters: affinity = rbf, gamma = 1
```
